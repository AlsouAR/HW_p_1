# Разработка программы тестирования на Python

Цель: Разработать консольную программу для проведения тестирования, которая позволяет загружать вопросы из файла, перемешивать их и варианты ответов, проводить тестирование пользователя, сохранять результаты в файл и предоставлять статистику по выполнению теста. Программа должна быть удобной для пользователя, с четкими инструкциями и немедленной обратной связью после каждого ответа.

```python
from google.colab import files
uploaded = files.upload()
import random
from datetime import datetime


def test_data(filename):
    test = []
    try:
        with open(filename, encoding='utf-8') as f:
            content = f.read().strip()
    except FileNotFoundError:
        raise FileNotFoundError(f"Ошибка: Файл не найден: {filename}")
    except Exception as e:
        raise IOError(f"Ошибка при чтении файла {filename}: {e}")
    mas1 = content.strip().split('\n\n\n')
    mas2 = [el.strip().split('\n') for el in mas1]
    for el in mas2:
        dict1 = {'answer_options': []}
        try:
            index_dot = el[0].find('.') + 2
            el[0] = el[0][index_dot:]
            dict1['questions'] = el[0]

            for i in range(1, len(el) - 1):
                index_dot = el[i].find('.') + 2
                el[i] = el[i][index_dot:]
                dict1['answer_options'].append(el[i])

            index_a = int(el[-1][7])
            dict1['answer'] = el[index_a]
            test.append(dict1)
        except Exception as e:
            raise ValueError(f"Ошибка при обработке элемента: {el}.Ошибка:{e}")
    return test


def show(data):
    print("=== Начинаем тестирование! ===")
    start = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    random.shuffle(data)

    total = len(data)
    correct = 0

    for idx, q in enumerate(data, 1):
        print(f"\nВопрос {idx}: {q['questions']}")

        random.shuffle(q['answer_options'])

        for num, opt in enumerate(q['answer_options'], 1):
            print(f"  {num}. {opt}")

        while True:
            try:
                user_ans = int(input("Ваш ответ (номер варианта): "))
                if 1 <= user_ans <= len(q['answer_options']):
                    break
                else:
                    print(f"Введите число от 1 до {len(q['answer_options'])}.")
            except ValueError:
                print("Пожалуйста, введите число.")

        correct_index = q['answer_options'].index(q['answer']) + 1
        if user_ans == correct_index:
            print("Правильно!")
            correct += 1
        else:
            print(f"Неверно. Правильный ответ: {correct_index}. {q['answer']}")

    print("\n=== Тест завершён ===")
    end = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    print(f"Правильно {correct} из {total}.")
    pct = correct / total * 100 if total else 0
    print(f"Процент правильных ответов: {pct:.2f}%")
    print("Результаты тестирования записаны в файл 'result.txt'")
    with open('result.txt', "w", encoding="utf-8") as file:
        file.write(f'''Время начала теста: {start}
Время окончания теста: {end}
Общее количество вопросов: {total}
Количество правильных ответов: {correct}
Процент правильных ответов: {pct}%''')


file_data = 'question.txt'
data = test_data(file_data)
show(data)
```
